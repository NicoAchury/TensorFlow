```python
# Librerias
import pandas as pd
import numpy as np
import datetime
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
# API de XM - Toda la informacion traida de la API es de carácter público.
import pydataxm as XM
import urllib.request
XM =  XM.ReadDB()
```

```python
# Último día disponible en la base de datos
Yesterday = datetime.date.today() - datetime.timedelta(days=1)
```

```python
# Traer información de la API de XM
ScarcityPrice = XM.request_data("PrecEscaAct",0,datetime.date(2018,1,1),Yesterday)
Wholesale = XM.request_data("PrecBolsNaci",0,datetime.date(2018,1,1),Yesterday)
HAdditions = XM.request_data("AporEner",0,datetime.date(2018,1,1),Yesterday)
```

```python
#Procesar información de precio de escasez
ScarcityPrice.drop(["Id"],axis=1,inplace=True)
ScarcityPrice["Date"] = pd.to_datetime(ScarcityPrice["Date"])
ScarcityPrice["Value"] = pd.to_numeric(ScarcityPrice["Value"])
ScarcityPrice.set_index("Date",inplace=True)
```

```python
#Procesar información de precio de bolsa
Wholesale.drop(["Id","Values_code"],axis=1,inplace=True)
Wholesale.iloc[:,0:24] = Wholesale.iloc[:,0:24].apply(pd.to_numeric)
Wholesale["Value"] = Wholesale.iloc[:,0:24].mean(axis=1)
Wholesale.drop(list(Wholesale.columns.values)[0:24],axis=1,inplace=True)
Wholesale["Date"] = pd.to_datetime(Wholesale["Date"])
Wholesale.set_index("Date",inplace=True)
```

```python
#Procesar aportes hidricos
HAdditions.drop(["Id"],axis=1,inplace=True)
HAdditions["Value"] = pd.to_numeric(HAdditions["Value"])/1e6
HAdditions["Date"] = pd.to_datetime(HAdditions["Date"])
HAdditions.set_index("Date",inplace=True)
```

```python
# Tipo de grafica a usar
plt.style.use("ggplot")
fig, ax1 = plt.subplots(figsize=(16,6))
#-----------------Aportes hídricos---------------
color="tab:blue"
ax1.set_ylabel("GWh", color=color)
HAdditions.plot(ax=ax1, label = "Aportes Hídricos", use_index=True, y="Value", kind="line", color=color)
ax1.tick_params(axis='y', labelcolor=color)
#-----------------Precio de bolsa----------------
ax2 = ax1.twinx()
color="tab:green"
ax2.set_ylabel("COP/kWh",color=color)
Wholesale.plot(ax=ax2,label = "Precio de Bolsa Promedio", use_index=True, y = "Value",kind="line", color=color)
ax2.tick_params(axis='y', labelcolor=color)
fig.tight_layout()
ax1.legend(loc='upper left',fontsize="large")
ax2.legend(loc='upper right',fontsize="large")
plt.title("Precio de Bolsa vs Aportes Hidrológicos")
ax1.set_xlabel("Fecha")
plt.show()
```

```python
#Graficar precio de bolsa vs precio de escasez
plt.figure(figsize=(16,6))
ax = plt.gca(ylabel="COP/kWh",xlabel="Date")
ScarcityPrice.plot(ax=ax,label="Precio de Escasez",use_index=True,y="Value",kind="line", color="tab:red")
Wholesale.plot(ax=ax, label="Precio de Bolsa Promedio",use_index=True,y="Value",kind="line", color="tab:green")
ax.set_xlabel("Fecha")
plt.title("Precio de Bolsa vs Precio de Escasez")
plt.show()
```

# ARIMA

Cálculo de parámetros p,d,q para un modelo ARIMA que permita realizar pronósticos de corto plazo para el precio de bolsa del mercado eléctrico en Colombia.

```python
#Identificar si la serie de tiempo tiene tendencia a través de la prueba aumentada de Dickey-Fuller
from statsmodels.tsa.stattools import adfuller
dftest = adfuller(Wholesale["Value"],autolag = "AIC")
Result = pd.Series(dftest[0:4], index = ["Test Statistic", "p-value", "Lags","# Observations"])
print(Result)   
```

Se rechaza la hipótesis por p-value alto y se considera que la serie de tiempo tiene una tendencia

### Descomposición de la serie de tiempo

```python
# Realizar descomposición de la serie de tiempo
from statsmodels.tsa.seasonal import seasonal_decompose
Trend = seasonal_decompose(Wholesale).trend
Sea = seasonal_decompose(Wholesale).seasonal
Resi = seasonal_decompose(Wholesale).resid
```

```python
# Graficar tendencia, estacionalidad y residual de la serie para el precio de bolsa.

fig, axes = plt.subplots(4, 1, sharey=False, figsize=(16,10))

Wholesale.plot(ax=axes[0],label="Precio de Bolsa Promedio",use_index=True,y="Value",kind="line",color="tab:green")
Trend.plot(ax=axes[1],label="Tendencia",use_index=True,y="Value",kind="line",color="tab:orange")
Sea.plot(ax=axes[2],label="Estacionalidad",use_index=True,y="Value",kind="line",color="tab:blue")
Resi.plot(ax=axes[3],label="Residual",use_index=True,y="Value",kind="line",color="tab:red")

for ax in axes:
    ax.set_ylabel("COP/kWh")
    ax.set_xlabel("Fecha")
    ax.legend(loc="best")

fig.tight_layout()
plt.show()
```

### Construir modelo ARIMA

Se utiliza PMDARIMA que permite calcular el mejor modelo ARIMA minimizando el Akaike Information Critera (AIC)

```python
# Dividir la serie entre Test y Train para el desarrollo del modelo
# Test = 0.2*Data
Train, Test = train_test_split(Wholesale.Value, shuffle=False, test_size=0.2)
```

```python
# Desarrollo del modelo de manera automatica a traves de Auto Arima. El modelo se ajusta con la 
# información de entrenamiento (Train) y se evalua con la información de prueba (Test)
import pmdarima as pm
ARIMA = pm.auto_arima(Train, start_p=1, start_q=1,
                         test='adf',
                         max_p=3, max_q=3, m=7,
                         start_P=0, seasonal=True,
                         d=None, trace=True,
                         error_action='ignore',  
                         suppress_warnings=True, 
                         stepwise=True)
```

```python
# Descripción del modelo
print(ARIMA.summary())
```

```python
#Realizar el pronóstico de la información de Prueba respecto al pronóstico del modelo.
# Se realiza un pronóstico de 1 periodo y luego se actualiza el modelo.
from sklearn.metrics import mean_squared_error
from pmdarima.metrics import smape

model = ARIMA
ForecastsInData = []
Confidence_Int_InData = []

# Función para pronosticar una unidad de tiempo (1 periodo). Intervalo de confianza de 95%
def forecast_one_step():
    fc, conf_int = model.predict(n_periods=1, return_conf_int=True, alpha=0.05)
    return (
        fc.tolist()[0],
        np.asarray(conf_int).tolist()[0])

#Realizar un número de pronósticos igual al tamaño de la serie Test para que puedan compararse
for x in Test:
    FC, CONF_INT = forecast_one_step()
    ForecastsInData.append(FC)
    Confidence_Int_InData.append(CONF_INT)
    
    # Actualizar el modelo por cada pronóstico hecho
    model.update(x)
```

```python
#Calcular metricas respecto a los datos de prueba y los pronóstico hechos por el modelo
from sklearn import metrics
print("Las metricas para los datos de prueba son: \n ")
print("Mean Absolute Error (MAE):      ", metrics.mean_absolute_error(Test,ForecastsInData))
print("Mean Squared Error (MSE):       ", metrics.mean_squared_error(Test,ForecastsInData))
print("Root Mean Squared Error (RMAE): ", np.sqrt(metrics.mean_squared_error(Test,ForecastsInData)))
```

```python
#Pronostico fuera del sample
# Se pronosticas 30 dias en el futuro
Days = 30
Day = Wholesale.index[-1]
T = []
for D in range(Days):
    Day = Day + datetime.timedelta(days=1)
    T.append(Day)
    
FC, CONF_INT = model.predict(n_periods=Days, return_conf_int=True, alpha=0.05)
FC=pd.Series(FC)
FC.index=T
FC.index.name="Date"
```

### Visualización de datos

```python
#Dataset para los datos de prueba
ForecastsInData = pd.Series(ForecastsInData)
ForecastsInData.index = Test.index
#--------------------------Ajuste del modelo SARIMA a los datos de prueba--------------
fig, axes = plt.subplots(2, 1, figsize=(16, 12))
Train.plot(ax=axes[0],color='tab:green',label="Datos Entrenamiento",kind="line")
Test.plot(ax=axes[0],color='tab:red',label="Datos Prueba",kind="line")
ForecastsInData.plot(ax=axes[0],color='tab:blue',label="Fit SARIMA",kind="line")
axes[0].legend(loc="upper left", prop={'size': 14})
axes[0].set_title("Ajuste del Modelo SARIMA")

#-------------------------Pronostico de 30 dias en el futuro SARIMA----------------------
Wholesale.Value.plot(ax=axes[1],color="tab:green",label="Precio de Bolsa",kind='line')
FC.plot(ax=axes[1],color="tab:red",label="Pronóstico",kind="line")
axes[1].fill_between(FC.index,CONF_INT[:,0],CONF_INT[:,1],alpha=0.4,color="tab:orange",
                     label="Intervalo de Confianza - 95%")
axes[1].set_xlim(pd.Timestamp(2020,6,1),T[-1])
axes[1].legend(loc="upper right", prop={'size': 14})
axes[1].set_title("Pronóstico e Intervalo de Confianza Precio de Bolsa - SARIMA")

for x in axes:
    x.set_ylabel("COP/kWh")
    x.set_xlabel("Fecha")

fig.tight_layout()
plt.show()
```

```python
#Calcular metricas respecto a los datos de prueba y los pronóstico hechos por el modelo
from sklearn import metrics
print("Las metricas para los datos de prueba son: \n ")
print("Mean Absolute Error (MAE):      ", round(metrics.mean_absolute_error(Test,ForecastsInData),2))
print("Mean Squared Error (MSE):       ", round(metrics.mean_squared_error(Test,ForecastsInData),2))
print("Root Mean Squared Error (RMAE): ", round(np.sqrt(metrics.mean_squared_error(Test,ForecastsInData)),2))
```

# Long Short Term Memory (LSTM) - Deep Learning

```python
# Importar librerias adicionales para LSTM
from sklearn.preprocessing import MinMaxScaler
from keras.models import Sequential
from keras.layers import Dense, LSTM
```

```python
#Escalamiento de datos
scaler = MinMaxScaler(feature_range=(0,1))
ScaleWholesale = scaler.fit_transform(Wholesale)
```

```python
#Obtener Train y Test data para el modelo. Se condisera Test=0.2*Serie de Tiempo
TrainDL, TestDL = train_test_split(ScaleWholesale, shuffle=False, test_size=0.2)
```

```python
#Estructuracion de la Ventana Temporal para datos de entrenamiento para el modelo
x_TrainDL=[]
y_TrainDL=[]
Vt=60

for i in range(Vt, len(TrainDL)):
    x_TrainDL.append(TrainDL[i-Vt:i,0])
    y_TrainDL.append(TrainDL[i,0])
    
x_TrainDL=np.array(x_TrainDL)
y_TrainDL=np.array(y_TrainDL)
```

```python
#Reshape
x_TrainDL = np.reshape(x_TrainDL,(x_TrainDL.shape[0],x_TrainDL.shape[1],1))
```

```python
#Consutruccion del modelo
LSTM_model = Sequential()
LSTM_model.add(LSTM(50,return_sequences=True, input_shape=(x_TrainDL.shape[1],1)))
LSTM_model.add(LSTM(50,return_sequences=False))
LSTM_model.add(Dense(25))
LSTM_model.add(Dense(25))
LSTM_model.add(Dense(1))
```

```python
#Compilar
LSTM_model.compile(optimizer="adam", loss="mean_squared_error")
```

```python
#Train
LSTM_model.fit(x_TrainDL,y_TrainDL,batch_size=4, epochs=10)
```

```python
#Estructuracion de la Ventana Temporal para datos de prueba para el modelo
TestDL = ScaleWholesale[len(TrainDL)-Vt:,:]

x_TestDL =[]
y_TestDL = scaler.inverse_transform(ScaleWholesale)[len(TrainDL):,:]

for i in range (Vt, len(TestDL)):
    x_TestDL.append(TestDL[i-Vt:i,0])
```

```python
#Reshape
x_TestDL = np.array(x_TestDL)
x_TestDL = np.reshape(x_TestDL,(x_TestDL.shape[0],x_TestDL.shape[1],1))
```

```python
#Pronostico para datos dentro de la muestra
ForecastsLSTM_OnSample = LSTM_model.predict(x_TestDL)
ForecastsLSTM_OnSample = scaler.inverse_transform(ForecastsLSTM_OnSample)

ForecastsLSTM_OnSample = pd.DataFrame(ForecastsLSTM_OnSample)
ForecastsLSTM_OnSample.index = Wholesale.iloc[len(TrainDL):,:].index
ForecastsLSTM_OnSample.columns = ["Value"]
```

```python
#Prediccion para datos fuera de la muestra - 30 dias a futuro
Days = 30
Window = ScaleWholesale[-Vt:,0]
ForecastLSTM_OutSample = np.array([])

for i in range(Days):
    Window = np.reshape(Window,(1,Vt,1))
    F = LSTM_model.predict(Window)
    ForecastLSTM_OutSample = np.append(ForecastLSTM_OutSample,F[0])
    Window = np.append(Window, F[0])
    Window = Window[-Vt:]  

ForecastLSTM_OutSample=np.reshape(ForecastLSTM_OutSample,(Days,1))
```

```python
# Estructuracion datos fuera de la muestra
ForecastLSTM_OutSample = scaler.inverse_transform(ForecastLSTM_OutSample)
ForecastLSTM_OutSample = pd.Series(np.reshape(ForecastLSTM_OutSample,(Days)))
ForecastLSTM_OutSample.index=T
```

```python
# Graficas

fig, axes = plt.subplots(2,1,figsize=(16,12))
#--------------------------Ajuste del modelo LSTM a los datos de prueba--------------
Wholesale.iloc[:len(TrainDL),:].plot(ax=axes[0], label="Datos Entrenamiento",
                                     use_index=True,y="Value",kind="line", color="tab:green")
Wholesale.iloc[len(TrainDL):,:].plot(ax=axes[0], label="Datos Prueba",
                                     use_index=True,y="Value",kind="line", color="tab:red")
ForecastsLSTM_OnSample.plot(ax=axes[0], label="Fit LSTM", use_index=True, y = "Value", kind="line", color="tab:purple")
axes[0].legend(loc="upper left", prop={'size': 14})
axes[0].set_title("Ajuste del Modelo LSTM")

#-------------------------Pronostico de 30 dias en el futuro LSTM----------------------

Wholesale.plot(ax=axes[1], label="Datos Entrenamiento",
                                     use_index=True,y="Value",kind="line", color="tab:green")
axes[1].set_title("Pronóstico e Intervalo de Confianza Precio de Bolsa - LSTM")
ForecastLSTM_OutSample.plot(ax=axes[1], label="Pronostico - LSTM", use_index=True, kind="line", color="tab:red")
axes[1].set_xlim(pd.Timestamp(2020,6,1),T[-1])
axes[1].legend(loc="upper left", prop={'size': 14})

for x in axes:
    x.set_ylabel("COP/kWh")
    x.set_xlabel("Fecha")

plt.tight_layout()
plt.show()
```

```python
#Metricas
print("Las metricas para los datos de prueba son: \n ")
print("Mean Absolute Error (MAE):      ", round(metrics.mean_absolute_error(y_TestDL,ForecastsLSTM_OnSample),2))
print("Mean Squared Error (MSE):       ", round(metrics.mean_squared_error(y_TestDL,ForecastsLSTM_OnSample),2))
print("Root Mean Squared Error (RMAE): ", round(np.sqrt(metrics.mean_squared_error(y_TestDL,ForecastsLSTM_OnSample)),2))
```
